```python
import pandas as pd
import re

# Load your combined dataset
csv_path = r"C:\Users\TTB\Desktop\Sugandha_Study\Sem3\GenAI\Project\dataset\combined_crisismmd_dataset.csv"
df = pd.read_csv(csv_path)

# Clean tweet text
def clean_tweet(text):
    if pd.isna(text):
        return ""
    
    text = text.strip()
    text = re.sub(r'^RT[\s]+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'www.\S+', '', text)
    text = text.encode('ascii', 'ignore').decode('ascii')
    text = re.sub(r'[^\w\s#]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = text.lower()
    return text

# Extract hashtags as comma-separated string
def extract_hashtag_string(text):
    if pd.isna(text):
        return ""
    hashtags = re.findall(r'#(\w+)', text)
    return ', '.join(hashtags)

# Create processed DataFrame
processed_df = pd.DataFrame()
processed_df['tweet_id'] = df['tweet_id'].astype(str)

# Extract only the image filename (not full path)
processed_df['image_id'] = df['image_path'].astype(str).apply(
    lambda x: x.split('/')[-1] if pd.notna(x) else ""
)

# Add raw tweet before cleaning
processed_df['raw_tweet_text'] = df['tweet_text']

# Add cleaned tweet text
processed_df['tweet_text'] = df['tweet_text'].apply(clean_tweet)

# Extract hashtags from cleaned tweet text
processed_df['tweet_hashtags'] = processed_df['tweet_text'].apply(extract_hashtag_string)

# Save final ground truth CSV
output_path = r"C:\Users\TTB\Desktop\Sugandha_Study\Sem3\GenAI\Project\dataset\groundtruth_crisis.csv"
processed_df.to_csv(output_path, index=False)

print(f"Final groundtruth_crisis.csv saved to:\n{output_path}")
print(f"--> Final shape: {processed_df.shape}")
processed_df.head()
```

# Creating Ground Truth using rule-based logic.

```python
import pandas as pd
import re
import spacy
from tqdm import tqdm

# ------------------ LOAD CSV ------------------
csv_path = r"C:\Users\TTB\Desktop\Sugandha_Study\Sem3\GenAI\Project\dataset\groundtruth_crisis_with_captions.csv"
df = pd.read_csv(csv_path)

# ------------------ DISTRESS KEYWORDS ------------------
distress_keywords = [
    # Medical emergencies
    "injured", "wounded", "hurt", "bleeding", "fatal", "fatalities", "casualties", "critical condition",
    "hospital", "ambulance", "can't breathe", "severely injured", "medical emergency", "urgent medical help",
    "severely hurt", "need medical help", "life-threatening", "emergency medical aid",
    
    # Rescue needs
    "trapped", "stuck", "collapsed house", "building collapsed", "trapped under debris",
    "can't move", "cannot move", "can't get out", "urgent rescue needed", "rescue needed urgently", 
    "stuck inside", "need rescue", "rescue needed", "trapped in building", "house collapsed", 
    "buried under rubble",
    
    # Missing people
    "missing", "lost", "can't find", "can't locate", "person missing", "missing person", 
    "people missing", "missing people", "family missing", "missing family", 
    "separated from family", "family separated", "lost contact", "lost my family", "missing relatives",
    "can't find family", "search and rescue",
    
    # Shelter, evacuation, abandoning
    "evacuate", "evacuation", "mass evacuation", "forced evacuation", "forced to evacuate", 
    "leave behind", "left behind", "left stranded", "forced to leave home", 
    "need shelter", "no shelter", "displaced", "lost our home", "homeless due to disaster",
    "need safe place", "evacuating area", "left home", "evacuated area", "abandoned home",
    "lost home", "home lost", "forced abandonment",
    
    # Panic and fear
    "panic", "panicking", "fear", "scared", "terrified", "chaos", "mayhem", "screaming",
    "people running", "running in fear", "stampede", "crowd panic",
    
    # Explicit help calls
    "help", "help me", "help us", "save me", "save us", "urgent help", "emergency help needed",
    "please rescue", "rescue needed now", "send help", "firefighters needed", "need urgent help",
    "please help", "help required", "someone help", "requesting urgent help"
]


# ------------------ CLEAN TEXT FUNCTION ------------------
def clean_text(text):
    text = str(text)
    text = re.sub(r"http\S+", "", text)  # remove URLs
    text = re.sub(r"#", "", text)        # remove hashtag symbol
    text = re.sub(r"@\w+", "", text)     # remove mentions
    return text.strip()

# ------------------ DETECT DISTRESS FUNCTION ------------------
def detect_distress(text):
    text = str(text).lower()
    if any(keyword in text for keyword in distress_keywords):
        return 1
    return 0

# ------------------ RECOMMEND ACTION FUNCTION ------------------
def recommend_action(distress, text):
    if distress == 0:
        return ""
    text = str(text).lower()
    if any(word in text for word in ["urgent", "emergency", "critical", "life-threatening"]):
        return "send immediate help"
    elif any(word in text for word in ["injured", "bleeding", "can't breathe", "severely hurt"]):
        return "send medical team"
    elif any(word in text for word in ["trapped", "stuck", "can't move", "urgent rescue"]):
        return "send rescue team"
    elif any(word in text for word in ["missing", "lost", "separated", "can't find"]):
        return "start missing person search"
    elif any(word in text for word in ["displaced", "evacuation", "need shelter", "lost our home"]):
        return "send evacuation and shelter support"
    elif any(word in text for word in ["help me", "help us", "save me", "save us", "please rescue"]):
        return "send security and trauma support"
    else:
        return "monitor situation"

# ------------------ LOCATION EXTRACTION FUNCTION ------------------
nlp = spacy.load("en_core_web_trf")  # Must be installed

def extract_location(text):
    doc = nlp(clean_text(text))
    locations = [ent.text for ent in doc.ents if ent.label_ in ["GPE", "LOC", "FAC"]]
    return ", ".join(locations) if locations else ""

# ------------------ APPLY TO DATAFRAME ------------------
tqdm.pandas()

df['distress'] = df['tweet_text'].progress_apply(detect_distress)
df['take_action'] = df.progress_apply(lambda row: recommend_action(row['distress'], row['tweet_text']), axis=1)
df['location'] = df['tweet_text'].progress_apply(extract_location)

# ------------------ SAVE OUTPUT ------------------
output_path = r"C:\Users\TTB\Desktop\Sugandha_Study\Sem3\GenAI\Project\dataset\groundtruth_crisis_labeled.csv"
df.to_csv(output_path, index=False)

print(f"\nLabeled dataset saved to:\n{output_path}")
print(f"--> Final shape: {df.shape}")
```

# Creating a well-balanced test sample from the groundtruth dataset to be used for validation of the system. (sample - rest split)

```python
import pandas as pd

# Assume your dataframe is loaded as df
# and it has 'disaster_type' and 'distress' columns

# Step 1: Initialize empty list to collect samples
final_samples = []

# Step 2: Disaster types
disaster_types = ['hurricane', 'wildfire', 'earthquake', 'flood', 'other']

# Step 3: For each disaster type
for d_type in disaster_types:
    df_type = df[df['disaster_type'] == d_type]
    
    # Subset distress=1 (positive) and distress=0 (negative)
    positive = df_type[df_type['distress'] == 1]
    negative = df_type[df_type['distress'] == 0]
    
    # We need at least 5 distress=1
    n_positive = min(5, len(positive))
    n_negative = 10 - n_positive  # Rest from distress=0
    
    if len(positive) < 5:
        print(f"⚠️ Warning: Only {len(positive)} positive samples available for {d_type}")
    
    # Sample
    sampled_positive = positive.sample(n=n_positive, random_state=42)
    sampled_negative = negative.sample(n=n_negative, random_state=42)
    
    # Combine for this disaster type
    sampled = pd.concat([sampled_positive, sampled_negative])
    
    # Shuffle to mix positives and negatives
    sampled = sampled.sample(frac=1, random_state=42).reset_index(drop=True)
    
    final_samples.append(sampled)

# Step 4: Combine all disaster types
sample_df = pd.concat(final_samples).reset_index(drop=True)

# Step 5: Get the rest of the data
rest_df = df.drop(sample_df.index)

# Step 6: Save to CSV
sample_df.to_csv('sample_dataset_balanced.csv', index=False)
rest_df.to_csv('rest_dataset.csv', index=False)

# Step 7: Check
print(f"✅ Sample dataset shape: {sample_df.shape}")
print(f"✅ Rest dataset shape: {rest_df.shape}")
print(f"✅ Sample dataset distribution:\n")
print(sample_df.groupby(['disaster_type', 'distress']).size())
```

# Creating ICR Triplets using Ground Truth dataset.

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
from IPython.display import display, HTML
```

```python
# Load splits
train_df = pd.read_csv("train_dataset.csv")
val_df = pd.read_csv("val_dataset.csv")
test_df = pd.read_csv("test_dataset.csv")

# Clean instruction (final version)
instruction = (
    "Classify this emergency tweet using the provided context.\n\n"
    "You must return the result in the following JSON format with these exact keys:\n"
    '- disaster_type: one of ["wildfire", "earthquake", "flood", "hurricane", "other"]\n'
    '- distress: either "distress" or "not distress"\n'
    '- location_info: extract the most specific location from the tweet (state and sub-location), or use \"unknown\"\n'
    '- action: one of ["send immediate help", "send medical team", "send rescue team", "start missing person search", "send evacuation help and shelter support", "send security and trauma support"]\n\n'
    "Return a valid JSON with all four fields."
)

# Function to format a row
def make_icr(row):
    tweet = str(row["tweet_text"]).strip()
    context = tweet  # Using tweet_text only

    response = {
        "disaster_type": str(row["disaster_type"]).strip().lower(),
        "distress": "distress" if int(row["distress"]) == 1 else "not distress",
        "location_info": f"{str(row['sub_location']).strip().lower()}, {str(row['state']).strip().lower()}",
        "action": str(row["take_action"]).strip().lower()
    }

    return {
        "instruction": instruction,
        "context": context,
        "response": json.dumps(response)
    }

# Apply formatting
train_icr = pd.DataFrame(train_df.apply(make_icr, axis=1).tolist())
val_icr = pd.DataFrame(val_df.apply(make_icr, axis=1).tolist())
test_icr = pd.DataFrame(test_df.apply(make_icr, axis=1).tolist())

# Save as CSVs
train_icr.to_csv("train_icr.csv", index=False)
val_icr.to_csv("val_icr.csv", index=False)
test_icr.to_csv("test_icr.csv", index=False)

print("ICR datasets saved: train_icr.csv, val_icr.csv, test_icr.csv")
```

```python
# Load ICR dataset
file_path = r"C:\Users\TTB\Desktop\Sugandha_Study\Sem3\GenAI\Project\project_jupyter\Fine_Tune\train_icr.csv"
icr_df = pd.read_csv(file_path)

def display_icr_triplets(df, num_samples=5):
    """
    Display ICR triplets in a formatted way
    
    Args:
        df: DataFrame containing the ICR triplets
        num_samples: Number of samples to display
    """
    # Show basic info
    print(f"Total number of ICR triplets: {len(df)}")
    print(f"Columns: {df.columns.tolist()}")
    print(f"\nDisplaying {num_samples} random samples:\n")
    
    # Get random samples
    samples = df.sample(num_samples)
    
    for i, (_, row) in enumerate(samples.iterrows()):
        print(f"{'='*30} Sample {i+1} {'='*30}")
        
        # Display instruction
        print("\n### INSTRUCTION:")
        print(row['instruction'])
        
        # Display context
        print("\n### CONTEXT:")
        print(row['context'])
        
        # Display response (parse JSON for better readability)
        print("\n### RESPONSE:")
        try:
            response_json = json.loads(row['response'])
            print(json.dumps(response_json, indent=4))
        except:
            print(row['response'])
        
        print("\n")

def analyze_icr_dataset(df):
    """
    Analyze the ICR dataset and show statistics
    """
    print(f"Dataset Analysis:")
    print(f"Total samples: {len(df)}")
    
    # Parse response JSONs
    responses = []
    for _, row in df.iterrows():
        try:
            response = json.loads(row['response'])
            responses.append(response)
        except:
            continue
    
    # Count disaster types
    disaster_counts = {}
    distress_counts = {}
    action_counts = {}
    
    for resp in responses:
        # Count disaster types
        if 'disaster_type' in resp:
            disaster_type = resp['disaster_type']
            disaster_counts[disaster_type] = disaster_counts.get(disaster_type, 0) + 1
        
        # Count distress
        if 'distress' in resp:
            distress = resp['distress']
            distress_counts[distress] = distress_counts.get(distress, 0) + 1
            
        # Count actions
        if 'action' in resp:
            action = resp['action']
            action_counts[action] = action_counts.get(action, 0) + 1
    
    # Display statistics
    print("\nDisaster Type Distribution:")
    for disaster, count in disaster_counts.items():
        percentage = (count / len(responses)) * 100
        print(f"  - {disaster}: {count} ({percentage:.1f}%)")
    
    print("\nDistress Distribution:")
    for distress, count in distress_counts.items():
        percentage = (count / len(responses)) * 100
        print(f"  - {distress}: {count} ({percentage:.1f}%)")
        
    print("\nAction Distribution:")
    for action, count in action_counts.items():
        percentage = (count / len(responses)) * 100
        print(f"  - {action}: {count} ({percentage:.1f}%)")
    
    # Visualize distributions
    plt.figure(figsize=(15, 5))
    
    # Disaster type distribution
    plt.subplot(1, 3, 1)
    plt.bar(disaster_counts.keys(), disaster_counts.values())
    plt.title('Disaster Type Distribution')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    
    # Distress distribution
    plt.subplot(1, 3, 2)
    plt.bar(distress_counts.keys(), distress_counts.values())
    plt.title('Distress Distribution')
    
    # Action distribution
    plt.subplot(1, 3, 3)
    plt.bar(action_counts.keys(), action_counts.values())
    plt.title('Action Distribution')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    
    plt.show()

# Display ICR triplets
display_icr_triplets(icr_df)

# Analyze dataset
analyze_icr_dataset(icr_df)
```

# Lora fine-tuning LLama using ICR triplets 

```python
import pandas as pd
import torch
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer,
    BitsAndBytesConfig,
    DataCollatorForLanguageModeling
)
from peft import get_peft_model, LoraConfig, TaskType, prepare_model_for_kbit_training

# Step 1: Load your ICR dataset
print("\nLoading dataset...")
df = pd.read_csv("train_icr.csv")
dataset = Dataset.from_pandas(df)

# Step 2: Format instruction+context+response into a single string
print("\nFormatting dataset...")
def format_prompt(example):
    return {
        "text": f"{example['instruction']}\n\nContext:\n{example['context']}\n\nAnswer: {example['response']}"
    }

formatted = dataset.map(format_prompt)

# Step 3: Load tokenizer and quantized model
model_name = "meta-llama/Llama-2-7b-chat-hf"
print("\nLoading tokenizer and model...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16
)

model = AutoModelForCausalLM.from_pretrained(model_name, quantization_config=bnb_config, device_map="auto")
model = prepare_model_for_kbit_training(model)

# Step 4: Apply LoRA
print("\nApplying LoRA config...")
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type=TaskType.CAUSAL_LM
)

model = get_peft_model(model, lora_config)

# Step 5: Tokenize the dataset
def tokenize(example):
    return tokenizer(example["text"], truncation=True, padding="max_length", max_length=512)

tokenized_dataset = formatted.map(tokenize)

# Step 6: Training setup
print("\nSetting up Trainer...")
args = TrainingArguments(
    output_dir="lora_llama_output",
    per_device_train_batch_size=4,
    gradient_accumulation_steps=2,
    num_train_epochs=15,
    logging_steps=20,
    save_strategy="epoch",
    learning_rate=2e-4,
    fp16=True,
    save_total_limit=2,
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized_dataset,
    tokenizer=tokenizer,
    data_collator=DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
)

# Step 7: Train
print("\nStarting LoRA fine-tuning...")
trainer.train()

# Step 8: Save model
print("\nSaving fine-tuned LoRA model...")
model.save_pretrained("lora_llama_output/final")
print("✅ LoRA fine-tuned model saved to: lora_llama_output/final")
```

```python
import pandas as pd
import torch
import numpy as np
import matplotlib.pyplot as plt
import json
import re
import seaborn as sns
import nltk
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftModel
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, classification_report, confusion_matrix
from tqdm import tqdm
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge_score import rouge_scorer

# Optional: For Jupyter environments
try:
    get_ipython().run_line_magic('matplotlib', 'inline')
except:
    pass

# Plot settings
plt.rcParams['figure.figsize'] = (12, 8)
plt.style.use('ggplot')

# Download required NLTK data
nltk.download('punkt', quiet=True)
```

```python
# Step 1: Load validation dataset
print("\nLoading validation dataset...")
df_val = pd.read_csv("val_icr.csv")
val_dataset = Dataset.from_pandas(df_val)

# Step 2: Load tokenizer and model with LoRA weights
model_name = "meta-llama/Llama-2-7b-chat-hf"
print("\nLoading tokenizer and model...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

# Load base model with quantization
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16
)
base_model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto"
)

# Load the fine-tuned LoRA model
print("\nLoading LoRA weights...")
model = PeftModel.from_pretrained(base_model, "lora_llama_output/final")
model.eval()

# Format validation examples same as training
def format_prompt(example):
    return f"{example['instruction']}\n\nContext:\n{example['context']}\n\nAnswer:"

# Step 3: Generate responses for validation set
print("\nGenerating responses for validation set...")
generated_responses = []
reference_responses = []

for i, example in enumerate(tqdm(val_dataset)):
    prompt = format_prompt(example)
    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")
    
    with torch.no_grad():
        generated_ids = model.generate(
            input_ids=input_ids,
            max_new_tokens=256,
            temperature=0.7,
            top_p=0.9,
            num_return_sequences=1,
            pad_token_id=tokenizer.eos_token_id
        )
    
    generated_text = tokenizer.decode(generated_ids[0][input_ids.shape[1]:], skip_special_tokens=True)
    generated_responses.append(generated_text)
    reference_responses.append(example["response"])

# Step 4: Parse JSON responses for classification tasks
print("\nParsing JSON responses...")

def extract_json(text):
    """Extract JSON from text, handling various formats"""
    # Try to find JSON pattern with curly braces
    json_pattern = re.compile(r'{.*?}', re.DOTALL)
    match = json_pattern.search(text)
    if match:
        json_text = match.group(0)
        try:
            return json.loads(json_text)
        except:
            pass
    
    # If not found, try to parse as a raw JSON
    try:
        return json.loads(text)
    except:
        return None

# Extract generated and reference classifications
gen_jsons = []
ref_jsons = []
successful_pairs = []

for i, (gen, ref) in enumerate(zip(generated_responses, reference_responses)):
    gen_json = extract_json(gen)
    ref_json = extract_json(ref)
    
    if gen_json and ref_json:
        gen_jsons.append(gen_json)
        ref_jsons.append(ref_json)
        successful_pairs.append(i)

print(f"Successfully parsed {len(gen_jsons)} out of {len(generated_responses)} response pairs")

# Step 5: Calculate Classification Metrics
print("\nCalculating classification metrics...")

# Initialize metrics for all tasks
tasks = ["disaster_type", "distress", "location_info", "action"]
task_metrics = {}

if len(gen_jsons) > 0:
    # For each classification task
    for task in tasks:
        print(f"\n--- {task.upper()} CLASSIFICATION METRICS ---")
        
        # Extract predictions and references
        y_true = [ref_json.get(task, "unknown") for ref_json in ref_jsons]
        y_pred = [gen_json.get(task, "unknown") for gen_json in gen_jsons]
        
        # Calculate accuracy
        accuracy = accuracy_score(y_true, y_pred)
        print(f"Accuracy: {accuracy:.4f}")
        
        # Generate classification report
        classes = sorted(list(set(y_true + y_pred)))
        report = classification_report(y_true, y_pred, output_dict=True, zero_division=0)
        print(classification_report(y_true, y_pred, zero_division=0))
        
        # Generate confusion matrix
        cm = confusion_matrix(y_true, y_pred, labels=classes)
        
        # Plot confusion matrix
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=classes, yticklabels=classes)
        plt.title(f'Confusion Matrix - {task}')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.tight_layout()
        plt.savefig(f'confusion_matrix_{task}.png')
        
        # Store metrics
        task_metrics[task] = {
            "accuracy": accuracy,
            "report": report
        }
    
    # Plot accuracy comparison
    plt.figure(figsize=(10, 6))
    accuracies = [task_metrics[task]["accuracy"] for task in tasks]
    plt.bar(tasks, accuracies, color='royalblue')
    plt.title('Classification Accuracy by Task')
    plt.ylabel('Accuracy')
    plt.ylim(0, 1)
    for i, v in enumerate(accuracies):
        plt.text(i, v + 0.02, f"{v:.3f}", ha='center')
    plt.tight_layout()
    plt.savefig('task_accuracies.png')
    
    # Save classification metrics
    with open('classification_metrics.json', 'w') as f:
        json.dump(task_metrics, f, indent=2)

# Step 6: Text Generation Metrics (BLEU & ROUGE)
print("\nCalculating text generation metrics...")

# 6.1 BLEU scores
bleu_scores = []
smoother = SmoothingFunction().method1

for gen, ref in zip(generated_responses, reference_responses):
    # Tokenize for BLEU
    gen_tokens = nltk.word_tokenize(gen.lower())
    ref_tokens = [nltk.word_tokenize(ref.lower())]
    
    try:
        # Calculate BLEU-4
        bleu = sentence_bleu(ref_tokens, gen_tokens, 
                            weights=(0.25, 0.25, 0.25, 0.25),
                            smoothing_function=smoother)
        bleu_scores.append(bleu)
    except:
        bleu_scores.append(0)

avg_bleu = np.mean(bleu_scores)
print(f"Average BLEU-4 score: {avg_bleu:.4f}")

# 6.2 ROUGE scores - using rouge_scorer instead of Rouge
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
rouge_scores = {"rouge1": [], "rouge2": [], "rougeL": []}

for gen, ref in zip(generated_responses, reference_responses):
    if len(gen.strip()) > 0 and len(ref.strip()) > 0:
        try:
            scores = scorer.score(ref, gen)
            rouge_scores["rouge1"].append(scores["rouge1"].fmeasure)
            rouge_scores["rouge2"].append(scores["rouge2"].fmeasure)
            rouge_scores["rougeL"].append(scores["rougeL"].fmeasure)
        except:
            # If error, add zeros
            rouge_scores["rouge1"].append(0)
            rouge_scores["rouge2"].append(0)
            rouge_scores["rougeL"].append(0)

avg_rouge1 = np.mean(rouge_scores["rouge1"])
avg_rouge2 = np.mean(rouge_scores["rouge2"])
avg_rougeL = np.mean(rouge_scores["rougeL"])

print(f"Average ROUGE-1 score: {avg_rouge1:.4f}")
print(f"Average ROUGE-2 score: {avg_rouge2:.4f}")
print(f"Average ROUGE-L score: {avg_rougeL:.4f}")

# Step 7: Visualizations for text generation metrics
print("\nGenerating visualizations...")

# 7.1 BLEU score distribution
plt.figure(figsize=(10, 6))
plt.hist(bleu_scores, bins=20, alpha=0.7, color='blue')
plt.axvline(avg_bleu, color='red', linestyle='dashed', linewidth=1)
plt.title('Distribution of BLEU Scores')
plt.xlabel('BLEU Score')
plt.ylabel('Frequency')
plt.savefig('bleu_distribution.png')

# 7.2 ROUGE scores comparison
plt.figure(figsize=(10, 6))
rouge_avgs = [avg_rouge1, avg_rouge2, avg_rougeL]
plt.bar(['ROUGE-1', 'ROUGE-2', 'ROUGE-L'], rouge_avgs, color=['skyblue', 'royalblue', 'navy'])
plt.title('Average ROUGE Scores')
plt.ylabel('Score')
plt.ylim(0, 1)
for i, v in enumerate(rouge_avgs):
    plt.text(i, v + 0.02, f"{v:.3f}", ha='center')
plt.savefig('rouge_scores.png')

# 7.3 BLEU and ROUGE scores by task
if successful_pairs:
    # Map successful pairs to their original indices
    task_bleu_scores = {task: [] for task in tasks}
    for idx in successful_pairs:
        # Get the original instruction
        instruction = val_dataset[idx]["instruction"]
        
        # Simple keyword matching to determine task type
        for task in tasks:
            if task.lower() in instruction.lower():
                task_bleu_scores[task].append(bleu_scores[idx])
    
    # Plot BLEU scores by task
    plt.figure(figsize=(10, 6))
    task_bleu_avgs = []
    valid_tasks = []
    
    for task in tasks:
        if task_bleu_scores[task]:
            task_bleu_avgs.append(np.mean(task_bleu_scores[task]))
            valid_tasks.append(task)
    
    if valid_tasks:
        plt.bar(valid_tasks, task_bleu_avgs, color='blue')
        plt.title('Average BLEU Score by Task Type')
        plt.ylabel('BLEU Score')
        plt.savefig('bleu_by_task.png')

# Step 8: Create detailed results CSV
print("\nSaving detailed results...")

# Create DataFrame with all results
results_df = pd.DataFrame({
    'original_instruction': [val_dataset[i]['instruction'] for i in range(len(val_dataset))],
    'original_context': [val_dataset[i]['context'] for i in range(len(val_dataset))],
    'reference_response': reference_responses,
    'generated_response': generated_responses,
    'bleu_score': bleu_scores
})

# Add ROUGE scores if available
if len(rouge_scores["rouge1"]) == len(bleu_scores):
    results_df['rouge1_score'] = rouge_scores["rouge1"]
    results_df['rouge2_score'] = rouge_scores["rouge2"] 
    results_df['rougeL_score'] = rouge_scores["rougeL"]

# Add parse success column
results_df['json_parse_success'] = [i in successful_pairs for i in range(len(generated_responses))]

# Save results
results_df.to_csv('detailed_validation_results.csv', index=False)

# Step 9: Summary of all metrics
print("\nSummary of all metrics:")
print(f"Total validation examples: {len(val_dataset)}")
print(f"Successfully parsed JSON pairs: {len(gen_jsons)} ({len(gen_jsons)/len(val_dataset)*100:.1f}%)")
print(f"Average BLEU score: {avg_bleu:.4f}")
print(f"Average ROUGE-L score: {avg_rougeL:.4f}")

if task_metrics:
    print("\nClassification task accuracies:")
    for task in tasks:
        if task in task_metrics:
            print(f"  {task}: {task_metrics[task]['accuracy']:.4f}")

print("\nEvaluation complete!")
```

```python
# path: emergency_response_rag_llama.py

import os
import faiss
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
from typing import List, Dict, Union, Any, Tuple
import re
import json
import logging
import time
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from peft import PeftModel, PeftConfig

# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)


class EmergencyResponseRAG:
    def __init__(self, index_path="document_faiss.index", dataset_path="rest_dataset_with_document.csv", embedding_model_name='all-mpnet-base-v2'):
        logger.info("Initializing Emergency Response RAG system...")
        self.index_path = index_path
        self.dataset_path = dataset_path
        self.retrieved_docs = []

        if not os.path.exists(index_path):
            logger.warning(f"FAISS index not found at {index_path}. Attempting to build it.")
            self.build_faiss_index(dataset_path, index_path, embedding_model_name)
        self.index = faiss.read_index(index_path)
        logger.info(f"Loaded FAISS index from {index_path}")

        self.df = pd.read_csv(dataset_path)
        logger.info(f"Loaded dataset from {dataset_path} with {len(self.df)} records")

        self.embedding_model = SentenceTransformer(embedding_model_name)
        logger.info(f"Loaded embedding model: {embedding_model_name}")

        self.model_path = r"C:\Users\TTB\Desktop\Sugandha_Study\Sem3\GenAI\Project\project_jupyter\Fine_Tune\lora_llama_output\final"
        peft_config = PeftConfig.from_pretrained(self.model_path)
        self.tokenizer = AutoTokenizer.from_pretrained(peft_config.base_model_name_or_path)
        base_model = AutoModelForCausalLM.from_pretrained(
            peft_config.base_model_name_or_path,
            torch_dtype=torch.float16,
            device_map="auto"
        )
        self.model = PeftModel.from_pretrained(base_model, self.model_path)
        self.pipe = pipeline("text-generation", model=self.model, tokenizer=self.tokenizer)
        logger.info("Loaded fine-tuned LoRA LLaMA model and tokenizer.")

        self.valid_disaster_types = ["wildfire", "earthquake", "flood", "hurricane", "other"]
        self.valid_distress_values = ["distress", "not distress"]
        self.valid_actions = ["monitor situation", "send evacuation help and shelter support", 
                              "send immediate help", "send medical team", "send rescue team", 
                              "send security and trauma support", "start missing person search"]
        
        # Common locations for emergency incidents
        self.known_locations = ["florida", "california", "texas", "new york", "louisiana", "oregon", 
                              "washington", "mexico", "puerto rico", "canada", "japan", "china",
                              "hawaii", "atlanta", "miami", "houston", "los angeles", "san francisco",
                              "chicago", "new orleans", "seattle", "boston", "phoenix", "philadelphia",
                              "puerto rico", "virgin islands", "north carolina", "south carolina",
                              "dominica", "dominican republic", "haiti", "cuba", "bahamas", "jamaica"]
        
        # Common words that might be mistaken for locations
        self.non_locations = ["hurricane", "earthquake", "flood", "wildfire", "the", "and", "in", 
                             "on", "at", "by", "from", "to", "with", "after", "before", "during",
                             "people", "person", "victims", "survivors", "officials", "authorities",
                             "emergency", "disaster", "crisis", "event", "incident", "situation", 
                             "killed", "injured", "missing", "dead", "alive", "rescued", "evacuation"]

    def build_faiss_index(self, dataset_path, index_path, model_name):
        df = pd.read_csv(dataset_path)
        if "document" not in df.columns:
            raise ValueError("The dataset must contain a 'document' column.")
        docs = df["document"].fillna("").tolist()
        model = SentenceTransformer(model_name)
        embeddings = model.encode(docs, convert_to_numpy=True, show_progress_bar=True)
        index = faiss.IndexFlatL2(embeddings.shape[1])
        index.add(embeddings)
        faiss.write_index(index, index_path)
        logger.info(f"FAISS index built and saved to: {index_path}")

    def clean_tweet(self, text: str) -> str:
        text = re.sub(r"http\S+|www\S+|https\S+", '', text)
        text = re.sub(r'\@\w+', '', text)
        text = re.sub(r'#', '', text)
        text = text.encode('ascii', 'ignore').decode('ascii')
        text = text.lower()
        text = re.sub(r'\s+', ' ', text).strip()
        return text

    def embed_text(self, text: str) -> np.ndarray:
        embedding = self.embedding_model.encode([text], convert_to_numpy=True)
        return embedding

    def retrieve_documents(self, query: str, k: int = 5) -> List[str]:
        """Retrieve relevant documents with increased k for better coverage"""
        query_vector = self.embed_text(query)
        distances, indices = self.index.search(query_vector, k)
        if len(indices[0]) == 0 or (indices[0][0] == -1):
            logger.warning("No relevant documents found in the index")
            return ["No relevant incident reports available."]
        
        # Return only unique documents
        unique_docs = []
        seen = set()
        for idx in indices[0]:
            if idx >= 0 and idx < len(self.df):  # Ensure index is valid
                doc = self.df.iloc[idx]['document']
                if doc not in seen and isinstance(doc, str) and len(doc.strip()) > 0:
                    unique_docs.append(doc)
                    seen.add(doc)
        
        return unique_docs[:3]  # Return top 3 unique documents

    def extract_json_from_text(self, text: str) -> Dict[str, Any]:
        """Extract valid JSON from text using robust pattern matching"""
        # Find JSON patterns - handles nested JSON better
        pattern = r'\{(?:[^{}]|(?:\{(?:[^{}]|(?:\{[^{}]*\}))*\}))*\}'
        matches = re.finditer(pattern, text)
        
        for match in matches:
            json_str = match.group(0)
            try:
                # Clean up any nested JSON issues
                cleaned_json = re.sub(r'\}\s*\{', '}, {', json_str)
                if cleaned_json.count('{') > cleaned_json.count('}'):
                    cleaned_json += '}'
                result = json.loads(cleaned_json)
                if isinstance(result, dict) and any(key in result for key in ["disaster_type", "distress", "location_info", "action"]):
                    return result
            except json.JSONDecodeError:
                continue
        
        return None

    def extract_location_from_tweet_first(self, tweet_text: str, docs: List[str]) -> Tuple[List[str], bool]:
        """Extract location with preference given to the tweet text first, then documents"""
        tweet_locations = []
        
        # First check for known locations in the tweet
        for loc in self.known_locations:
            if loc.lower() in tweet_text.lower():
                # Don't add if it's part of a larger word
                if re.search(r'\b' + re.escape(loc) + r'\b', tweet_text.lower()):
                    tweet_locations.append(loc.title())
        
        # Also check for any capitalized words that might be locations not in our list
        # This will help catch places like "Dominica" that aren't in the known_locations list
        potential_locations = re.findall(r'\b([A-Z][a-z]+)\b', tweet_text)
        for loc in potential_locations:
            if loc.lower() not in self.non_locations and loc.title() not in tweet_locations:
                tweet_locations.append(loc.title())
        
        # Check for location words preceded by "in"
        in_location_match = re.findall(r'\bin\s+(\w+)', tweet_text.lower())
        for loc in in_location_match:
            if loc not in self.non_locations and loc.title() not in tweet_locations:
                tweet_locations.append(loc.title())
        
        # If we found locations in the tweet, return those with flag indicating tweet source
        if tweet_locations:
            return tweet_locations, True
        
        # Otherwise, extract from documents
        doc_locations = []
        for doc in docs:
            # Search for "In [Location]" pattern
            loc_match = re.search(r'In ([^,]+)', doc)
            if loc_match and loc_match.group(1).lower() != "unknown location":
                loc = loc_match.group(1).strip()
                if "unknown" not in loc.lower():
                    doc_locations.append(loc)
            
            # Also look for known location names
            for loc in self.known_locations:
                if loc.lower() in doc.lower():
                    # Don't add if it's part of a larger word
                    if re.search(r'\b' + re.escape(loc) + r'\b', doc.lower()):
                        doc_locations.append(loc.title())
        
        return list(set(doc_locations)), False  # Remove duplicates and flag as document source

    def extract_disaster_types_from_tweet_first(self, tweet_text: str, docs: List[str]) -> Tuple[List[str], bool]:
        """Extract disaster types with preference given to the tweet text first"""
        tweet_disasters = []
        
        # First check tweet for disaster types
        for disaster_type in self.valid_disaster_types:
            if disaster_type.lower() in tweet_text.lower():
                tweet_disasters.append(disaster_type)
        
        # If found in tweet, return those
        if tweet_disasters:
            return tweet_disasters, True
            
        # Otherwise, check documents
        doc_disasters = []
        for doc in docs:
            for disaster_type in self.valid_disaster_types:
                if disaster_type.lower() in doc.lower():
                    doc_disasters.append(disaster_type)
        
        return list(set(doc_disasters)), False  # Remove duplicates and flag as document source

    def build_system_message(self) -> str:
        """Build a clear system message with improved instructions"""
        return """You are an emergency response assistant that specializes in analyzing social media posts about disasters.
Your task is to extract key information from tweets about emergencies and return a structured JSON analysis.

You must focus on information from both the tweet and the provided documents.

Important instructions:
1. Pay careful attention to location information in the documents
2. Determine the appropriate disaster type and action based on context
3. Return ONLY a JSON object with no additional text or commentary
4. Ensure JSON output is properly formatted

Return your analysis with these exact keys:
- disaster_type: One of [wildfire, earthquake, flood, hurricane, other]
- distress: Either 'distress' or 'not distress'
- location_info: Exact location from documents or 'unknown'
- action: One of [monitor situation, send evacuation help and shelter support, send immediate help, send medical team, send rescue team, send security and trauma support, start missing person search]
"""

    def build_user_message(self, tweet_text: str, retrieved_docs: List[str], 
                          locations: List[str], location_in_tweet: bool,
                          disaster_types: List[str], disaster_in_tweet: bool) -> str:
        """Build user message with explicit highlighting of key information"""
        message = f"Tweet to analyze: \"{tweet_text}\"\n\nRelevant emergency documents:\n"
        
        for idx, doc in enumerate(retrieved_docs, 1):
            message += f"Document {idx}: {doc}\n"
        
        # Explicitly highlight extracted information
        if locations:
            if location_in_tweet:
                message += f"\nLocation mentioned in tweet: {', '.join(locations)}\n"
            else:
                message += f"\nLocations from similar incidents: {', '.join(locations)}\n"
        
        if disaster_types:
            if disaster_in_tweet:
                message += f"Disaster type mentioned in tweet: {', '.join(disaster_types)}\n"
            else:
                message += f"Disaster types from similar incidents: {', '.join(disaster_types)}\n"
        
        message += """
Based on the tweet and retrieved documents above, provide a JSON with these fields:

{
  "disaster_type": select from [wildfire, earthquake, flood, hurricane, other],
  "distress": select from [distress, not distress],
  "location_info": extract the specific location from the tweet if available, otherwise from documents (use the actual location name instead of "unknown" if available),
  "action": select from [monitor situation, send evacuation help and shelter support, send immediate help, send medical team, send rescue team, send security and trauma support, start missing person search]
}

Return ONLY a valid JSON object.
"""
        return message

    def validate_output(self, result: Dict[str, Any], 
                       locations: List[str], location_in_tweet: bool, 
                       disaster_types: List[str], disaster_in_tweet: bool) -> Dict[str, Any]:
        """Validate and enhance model output with extracted information"""
        # Ensure all required fields are present
        for key in ["disaster_type", "distress", "location_info", "action"]:
            if key not in result:
                result[key] = "unknown"
        
        # Clean and enhance location information - FIXED VERSION
        if location_in_tweet and locations:
            # If location was found in tweet, ALWAYS use it
            result["location_info"] = locations[0]
        elif result["location_info"] == "unknown" or "unknown" in result["location_info"].lower():
            # Only use document locations if we don't have tweet locations
            if locations and not location_in_tweet:
                # Here we're keeping "unknown" for consistency when location not in tweet
                result["location_info"] = "unknown"
        elif isinstance(result["location_info"], str) and "," in result["location_info"]:
            # Handle comma-separated locations, remove "unknown"
            parts = [p.strip() for p in result["location_info"].split(",")]
            valid_parts = [p for p in parts if p.lower() != "unknown"]
            if valid_parts:
                result["location_info"] = valid_parts[0]
        
        # Validate disaster_type
        if result["disaster_type"].lower() not in self.valid_disaster_types:
            if disaster_types:
                result["disaster_type"] = disaster_types[0]
            else:
                result["disaster_type"] = "other"
        
        # Validate distress
        if result["distress"].lower() not in self.valid_distress_values:
            # If tweet mentions deaths, injuries, or missing people, it's likely distress
            tweet_text = result.get("tweet_text", "").lower()
            if any(word in tweet_text for word in ["killed", "dead", "died", "death", "missing", "injured", "trapped", "victims", "casualties"]):
                result["distress"] = "distress"
            else:
                result["distress"] = "not distress"
        
        # Map action to valid actions
        if result["action"] == "unknown" or result["action"] not in self.valid_actions:
            # Try to map to a valid action based on context
            tweet_text = result.get("tweet_text", "").lower()
            context_text = ' '.join(self.retrieved_docs).lower() if self.retrieved_docs else ""
            combined_text = tweet_text + " " + context_text
            
            # Default to "monitor situation" for non-distress situations
            action = "monitor situation"
            
            # Check for specific keywords to determine action
            if "missing" in combined_text or "search" in combined_text:
                action = "start missing person search"
            elif "evacuation" in combined_text or "shelter" in combined_text:
                action = "send evacuation help and shelter support"
            elif "medical" in combined_text or "injury" in combined_text or "injuries" in combined_text:
                action = "send medical team"
            elif "rescue" in combined_text or "trapped" in combined_text:
                action = "send rescue team"
            elif "security" in combined_text or "trauma" in combined_text:
                action = "send security and trauma support"
            elif "immediate help" in combined_text or "emergency" in combined_text:
                action = "send immediate help"
            
            # If tweet mentions deaths/casualties, likely need search and rescue
            if "killed" in tweet_text or "dead" in tweet_text or "died" in tweet_text or "death" in tweet_text:
                action = "start missing person search"
                
            result["action"] = action
        
        return result

    def generate_analysis(self, tweet_text: str, retrieved_docs: List[str],
                         locations: List[str], location_in_tweet: bool,
                         disaster_types: List[str], disaster_in_tweet: bool) -> Dict[str, Any]:
        """Generate emergency analysis with improved extraction and validation"""
        try:
            # Build structured prompt
            system_message = self.build_system_message()
            user_message = self.build_user_message(tweet_text, retrieved_docs, 
                                                 locations, location_in_tweet,
                                                 disaster_types, disaster_in_tweet)
            
            # Generate analysis
            prompt = f"{system_message}\n\n{user_message}"
            outputs = self.pipe(prompt, max_new_tokens=300, temperature=0.1, do_sample=False)
            content = outputs[0]["generated_text"]
            
            # Log raw output for debugging
            print("\n--- Raw LLaMA Output ---")
            print(content)
            print("------------------------\n")
            
            # Extract and parse JSON
            json_result = self.extract_json_from_text(content)
            
            if json_result:
                # Add tweet text for context in validation
                json_result["tweet_text"] = tweet_text
                # IMPORTANT: Always validate output with our extracted information
                return self.validate_output(json_result, locations, location_in_tweet, 
                                         disaster_types, disaster_in_tweet)
            else:
                # Fall back to rule-based extraction if JSON parsing fails
                result = self._get_default_response()
                result["tweet_text"] = tweet_text
                
                # Explicitly handle locations from tweet
                if locations and location_in_tweet:
                    result["location_info"] = locations[0]
                    
                if disaster_types:
                    result["disaster_type"] = disaster_types[0]
                
                return self.validate_output(result, locations, location_in_tweet, 
                                         disaster_types, disaster_in_tweet)

        except Exception as e:
            logger.error(f"Error in generate_analysis: {e}")
            # Create default response with extracted information
            result = self._get_default_response()
            result["tweet_text"] = tweet_text
            return self.validate_output(result, locations, location_in_tweet, 
                                     disaster_types, disaster_in_tweet)

    def _get_default_response(self) -> Dict[str, Any]:
        """Default fallback response"""
        return {
            "disaster_type": "other",
            "distress": "not distress",
            "location_info": "unknown",
            "action": "monitor situation"
        }

    def process_tweet(self, tweet_text: str) -> Dict[str, Any]:
        """Process a tweet and generate emergency response analysis"""
        try:
            start_time = time.time()
            
            # Clean tweet text
            cleaned_tweet = self.clean_tweet(tweet_text)
            logger.info(f"Cleaned tweet: {cleaned_tweet}")
            
            # Store original tweet for pattern matching
            original_tweet = tweet_text
            
            # Retrieve relevant documents
            self.retrieved_docs = self.retrieve_documents(cleaned_tweet, k=5)
            
            print("\n--- Retrieved Documents ---")
            for i, doc in enumerate(self.retrieved_docs, 1):
                print(f"Doc {i}: {doc}")
            print("---------------------------\n")
            
            # Extract locations with preference to tweet - use original tweet for capitalization
            locations, location_in_tweet = self.extract_location_from_tweet_first(original_tweet, self.retrieved_docs)
            
            # Extract disaster types with preference to tweet
            disaster_types, disaster_in_tweet = self.extract_disaster_types_from_tweet_first(cleaned_tweet, self.retrieved_docs)
            
            # Generate analysis
            result = self.generate_analysis(cleaned_tweet, self.retrieved_docs, 
                                         locations, location_in_tweet,
                                         disaster_types, disaster_in_tweet)
            
            # Ensure the location is correctly used from tweet if available
            if location_in_tweet and locations and result["location_info"] == "unknown":
                result["location_info"] = locations[0]
            
            # Calculate processing time
            duration = time.time() - start_time
            
            # Return complete result
            return {
                "input_tweet": tweet_text,
                "cleaned_tweet": cleaned_tweet,
                "retrieved_docs": self.retrieved_docs,
                "extracted_locations": locations,
                "location_in_tweet": location_in_tweet,
                "extracted_disaster_types": disaster_types,
                "disaster_in_tweet": disaster_in_tweet,
                "analysis": result,
                "processing_time": duration
            }
        except Exception as e:
            logger.error(f"Pipeline error: {e}")
            return {
                "input_tweet": tweet_text,
                "error": str(e),
                "analysis": self._get_default_response()
            }


if __name__ == "__main__":
    print("Emergency Tweet Response System (RAG + Local LLaMA)")
    print("Type 'exit' to quit.\n")

    try:
        rag_system = EmergencyResponseRAG()
        while True:
            tweet = input("Enter a new tweet: ")
            if tweet.lower() == "exit":
                break

            result = rag_system.process_tweet(tweet)

            print("\n=== Emergency Response Analysis ===")
            print(f"Tweet: {result['input_tweet']}")
            print("\nAnalysis:")
            if "error" in result:
                print(f"Error: {result['error']}")
            else:
                # Print extracted information for debugging
                if "extracted_locations" in result:
                    if result["location_in_tweet"]:
                        print(f"✓ Locations found in tweet: {result['extracted_locations']}")
                    elif result["extracted_locations"]:
                        print(f"ℹ Locations from similar incidents: {result['extracted_locations']}")
                    else:
                        print("No locations found in tweet or documents")
                
                if "extracted_disaster_types" in result:
                    if result["disaster_in_tweet"]:
                        print(f"✓ Disaster types found in tweet: {result['extracted_disaster_types']}")
                    elif result["extracted_disaster_types"]:
                        print(f"ℹ Disaster types from similar incidents: {result['extracted_disaster_types']}")
                    else:
                        print("No disaster types found in tweet or documents")
                
                a = result["analysis"]
                print(f"\nDisaster Type: {a['disaster_type']}")
                print(f"Distress Level: {a['distress']}")
                
                # Display location with source indicator
                if "location_in_tweet" in result and result["location_in_tweet"] and a["location_info"] != "unknown":
                    print(f"Location: {a['location_info']} (mentioned in tweet)")
                elif a["location_info"] != "unknown":
                    print(f"Location: {a['location_info']} (inferred from similar incidents)")
                else:
                    print(f"Location: unknown")
                
                print(f"Recommended Action: {a['action']}")
                print(f"Processing Time: {result['processing_time']:.2f} sec")
            print("=" * 50)

    except KeyboardInterrupt:
        print("\nInterrupted by user.")
    except Exception as e:
        print(f"\nAn error occurred: {e}")
    finally:
        print("\nEmergency Response System shutdown complete.")
```

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
import time
import json
import logging
import re
from typing import Dict, List, Any
# For metrics calculation
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge_score import rouge_scorer
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score, precision_score, recall_score

# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# Initialize Rouge scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

class RAGSystemEvaluator:
    def __init__(self, rag_system, sample_dataset_path="sample_dataset_balanced.csv", 
                 output_path="evaluation_results"):
        """Initialize the evaluator with dataset and RAG system"""
        self.sample_df = pd.read_csv(sample_dataset_path)
        print(f"Loaded sample dataset from {sample_dataset_path} with {len(self.sample_df)} samples")
        
        # Use the provided RAG system
        self.rag_system = rag_system
        self.output_path = output_path
        self.smoothing = SmoothingFunction().method1
        self.results = []
        self.sample_results = []  # Store complete sample results for display
        
    def calculate_bleu(self, reference: str, hypothesis: str) -> float:
        """Calculate BLEU score between reference and hypothesis"""
        if not reference or not hypothesis:
            return 0.0
            
        reference_tokens = reference.lower().split()
        hypothesis_tokens = hypothesis.lower().split()
        
        # BLEU requires a list of references
        return sentence_bleu([reference_tokens], hypothesis_tokens, 
                            smoothing_function=self.smoothing)
    
    def calculate_rouge(self, reference: str, hypothesis: str) -> Dict[str, Dict[str, float]]:
        """Calculate ROUGE scores between reference and hypothesis"""
        if not reference or not hypothesis:
            return {
                'rouge1': {'precision': 0.0, 'recall': 0.0, 'fmeasure': 0.0},
                'rouge2': {'precision': 0.0, 'recall': 0.0, 'fmeasure': 0.0},
                'rougeL': {'precision': 0.0, 'recall': 0.0, 'fmeasure': 0.0}
            }
            
        try:
            # Rouge requires non-empty strings
            if len(reference.strip()) == 0 or len(hypothesis.strip()) == 0:
                return {
                    'rouge1': {'precision': 0.0, 'recall': 0.0, 'fmeasure': 0.0},
                    'rouge2': {'precision': 0.0, 'recall': 0.0, 'fmeasure': 0.0},
                    'rougeL': {'precision': 0.0, 'recall': 0.0, 'fmeasure': 0.0}
                }
                
            scores = scorer.score(reference, hypothesis)
            
            # Convert the object to a dict with the format expected
            result = {}
            for key, value in scores.items():
                result[key] = {
                    'precision': value.precision,
                    'recall': value.recall,
                    'fmeasure': value.fmeasure
                }
                
            return result
        except Exception as e:
            print(f"Error calculating ROUGE: {e}")
            return {
                'rouge1': {'precision': 0.0, 'recall': 0.0, 'fmeasure': 0.0},
                'rouge2': {'precision': 0.0, 'recall': 0.0, 'fmeasure': 0.0},
                'rougeL': {'precision': 0.0, 'recall': 0.0, 'fmeasure': 0.0}
            }
    
    def calculate_exact_match(self, reference: str, hypothesis: str) -> float:
        """Calculate exact match score (1.0 if exact match, 0.0 otherwise)"""
        return 1.0 if reference.lower().strip() == hypothesis.lower().strip() else 0.0
        
    def evaluate_sample(self, row: pd.Series) -> Dict[str, Any]:
        """Evaluate a single sample from the dataset"""
        tweet_text = row['tweet_text']
        
        # Ground truth values
        gt_disaster_type = str(row['disaster_type']).lower() if not pd.isna(row['disaster_type']) else "unknown"
        gt_distress = "distress" if row['distress'] == 1 else "not distress"
        gt_location = str(row['state']).lower() if not pd.isna(row['state']) else "unknown"
        if not pd.isna(row['sub_location']) and str(row['sub_location']).strip():
            gt_location += f" ({str(row['sub_location']).lower()})"
        gt_action = str(row['take_action']).lower() if not pd.isna(row['take_action']) else "unknown"
        
        # Get RAG system prediction
        print(f"Processing: {tweet_text[:50]}...")
        start_time = time.time()
        prediction = self.rag_system.process_tweet(tweet_text)
        processing_time = time.time() - start_time
        
        pred_analysis = prediction['analysis']
        pred_disaster_type = str(pred_analysis['disaster_type']).lower()
        pred_distress = str(pred_analysis['distress']).lower()
        pred_location = str(pred_analysis['location_info']).lower()
        pred_action = str(pred_analysis['action']).lower()
        
        # Calculate metrics for each field
        disaster_type_bleu = self.calculate_bleu(gt_disaster_type, pred_disaster_type)
        disaster_type_exact = self.calculate_exact_match(gt_disaster_type, pred_disaster_type)
        disaster_type_rouge = self.calculate_rouge(gt_disaster_type, pred_disaster_type)
        
        distress_bleu = self.calculate_bleu(gt_distress, pred_distress)
        distress_exact = self.calculate_exact_match(gt_distress, pred_distress)
        distress_rouge = self.calculate_rouge(gt_distress, pred_distress)
        
        location_bleu = self.calculate_bleu(gt_location, pred_location)
        location_exact = self.calculate_exact_match(gt_location, pred_location)
        location_rouge = self.calculate_rouge(gt_location, pred_location)
        
        action_bleu = self.calculate_bleu(gt_action, pred_action)
        action_exact = self.calculate_exact_match(gt_action, pred_action)
        action_rouge = self.calculate_rouge(gt_action, pred_action)
        
        # Combine all results
        result = {
            'tweet_id': row['tweet_id'] if 'tweet_id' in row else 'unknown',
            'tweet_text': tweet_text,
            
            'gt_disaster_type': gt_disaster_type,
            'pred_disaster_type': pred_disaster_type,
            'disaster_type_bleu': disaster_type_bleu,
            'disaster_type_exact': disaster_type_exact,
            'disaster_type_rouge_1': disaster_type_rouge['rouge1']['fmeasure'],
            'disaster_type_rouge_2': disaster_type_rouge['rouge2']['fmeasure'],
            'disaster_type_rouge_l': disaster_type_rouge['rougeL']['fmeasure'],
            
            'gt_distress': gt_distress,
            'pred_distress': pred_distress,
            'distress_bleu': distress_bleu,
            'distress_exact': distress_exact,
            'distress_rouge_1': distress_rouge['rouge1']['fmeasure'],
            'distress_rouge_2': distress_rouge['rouge2']['fmeasure'],
            'distress_rouge_l': distress_rouge['rougeL']['fmeasure'],
            
            'gt_location': gt_location,
            'pred_location': pred_location,
            'location_bleu': location_bleu,
            'location_exact': location_exact,
            'location_rouge_1': location_rouge['rouge1']['fmeasure'],
            'location_rouge_2': location_rouge['rouge2']['fmeasure'],
            'location_rouge_l': location_rouge['rougeL']['fmeasure'],
            
            'gt_action': gt_action,
            'pred_action': pred_action,
            'action_bleu': action_bleu,
            'action_exact': action_exact,
            'action_rouge_1': action_rouge['rouge1']['fmeasure'],
            'action_rouge_2': action_rouge['rouge2']['fmeasure'],
            'action_rouge_l': action_rouge['rougeL']['fmeasure'],
            
            'processing_time': processing_time,
            'retrieved_docs': prediction.get('retrieved_docs', [])
        }
        
        # Store additional data for complete sample display
        sample_result = {
            'tweet_id': row['tweet_id'] if 'tweet_id' in row else 'unknown',
            'tweet_text': tweet_text,
            'ground_truth': {
                'disaster_type': gt_disaster_type,
                'distress': gt_distress,
                'location': gt_location,
                'action': gt_action
            },
            'prediction': {
                'disaster_type': pred_disaster_type,
                'distress': pred_distress,
                'location': pred_location,
                'action': pred_action
            },
            'metrics': {
                'disaster_type': {
                    'bleu': disaster_type_bleu,
                    'exact_match': disaster_type_exact,
                    'rouge_1': disaster_type_rouge['rouge1']['fmeasure'],
                    'rouge_l': disaster_type_rouge['rougeL']['fmeasure']
                },
                'distress': {
                    'bleu': distress_bleu,
                    'exact_match': distress_exact,
                    'rouge_1': distress_rouge['rouge1']['fmeasure'],
                    'rouge_l': distress_rouge['rougeL']['fmeasure']
                },
                'location': {
                    'bleu': location_bleu,
                    'exact_match': location_exact,
                    'rouge_1': location_rouge['rouge1']['fmeasure'],
                    'rouge_l': location_rouge['rougeL']['fmeasure']
                },
                'action': {
                    'bleu': action_bleu,
                    'exact_match': action_exact,
                    'rouge_1': action_rouge['rouge1']['fmeasure'],
                    'rouge_l': action_rouge['rougeL']['fmeasure']
                }
            },
            'full_prediction': prediction
        }
        
        # If there are locations extracted, save them
        if 'extracted_locations' in prediction:
            result['extracted_locations'] = prediction['extracted_locations']
            result['location_in_tweet'] = prediction.get('location_in_tweet', False)
            sample_result['extracted_locations'] = prediction['extracted_locations']
            sample_result['location_in_tweet'] = prediction.get('location_in_tweet', False)
        
        # Add sample result to list
        self.sample_results.append(sample_result)
        
        print(f"Processed in {processing_time:.2f} seconds")
        return result
        
    def evaluate_all_samples(self, max_samples=None):
        """Evaluate all samples in the dataset"""
        if max_samples:
            sample_df = self.sample_df.head(max_samples)
            print(f"Starting evaluation of {len(sample_df)} samples (limited to {max_samples})...")
        else:
            sample_df = self.sample_df
            print(f"Starting evaluation of all {len(sample_df)} samples...")
        
        # Use tqdm for progress bar in notebook
        for i, row in tqdm(sample_df.iterrows(), total=len(sample_df), desc="Evaluating"):
            try:
                result = self.evaluate_sample(row)
                self.results.append(result)
                
                # After every few samples, print current stats
                if len(self.results) % 5 == 0:
                    current_metrics = self.calculate_metrics_so_far()
                    print(f"\nInterim metrics after {len(self.results)} samples:")
                    print(f"Disaster Type Accuracy: {current_metrics.get('disaster_type_accuracy', 0):.4f}")
                    print(f"Distress Detection Accuracy: {current_metrics.get('distress_accuracy', 0):.4f}")
                    
            except Exception as e:
                print(f"Error processing sample {i}: {e}")
                
        # Calculate aggregate metrics
        metrics = self.calculate_aggregate_metrics()
        
        # Save results
        self.save_results()
        
        # Convert results to DataFrame
        results_df = pd.DataFrame(self.results)
        
        return results_df, metrics
    
    def display_sample_results(self, num_samples=5):
        """Display detailed results for a number of sample tweets"""
        if not self.sample_results:
            print("No sample results to display")
            return
            
        # Select samples to display
        if num_samples >= len(self.sample_results):
            samples_to_display = self.sample_results
        else:
            # Use a mix of correct and incorrect samples if possible
            results_df = pd.DataFrame(self.results)
            if 'disaster_type_exact' in results_df.columns:
                # Get indices of correct and incorrect predictions
                correct_indices = results_df[results_df['disaster_type_exact'] == 1.0].index.tolist()
                incorrect_indices = results_df[results_df['disaster_type_exact'] == 0.0].index.tolist()
                
                # Select a balanced mix
                num_correct = min(num_samples // 2, len(correct_indices))
                num_incorrect = min(num_samples - num_correct, len(incorrect_indices))
                
                if num_correct > 0:
                    correct_samples = np.random.choice(correct_indices, num_correct, replace=False)
                else:
                    correct_samples = []
                
                if num_incorrect > 0:
                    incorrect_samples = np.random.choice(incorrect_indices, num_incorrect, replace=False)
                else:
                    incorrect_samples = []
                
                # Combine and ensure we have enough samples
                selected_indices = list(correct_samples) + list(incorrect_samples)
                if len(selected_indices) < num_samples:
                    remaining = min(num_samples - len(selected_indices), len(self.sample_results) - len(selected_indices))
                    remaining_indices = list(set(range(len(self.sample_results))) - set(selected_indices))
                    if remaining > 0 and remaining_indices:
                        selected_indices.extend(np.random.choice(remaining_indices, remaining, replace=False))
                
                samples_to_display = [self.sample_results[i] for i in selected_indices]
            else:
                # Random selection if we can't determine correct/incorrect
                indices = np.random.choice(len(self.sample_results), num_samples, replace=False)
                samples_to_display = [self.sample_results[i] for i in indices]
        
        print(f"\n===== Sample Results ({len(samples_to_display)} samples) =====\n")
        
        for i, sample in enumerate(samples_to_display, 1):
            print(f"Sample {i}:")
            print(f"Tweet: {sample['tweet_text']}")
            print("\nGround Truth:")
            print(f"  Disaster Type: {sample['ground_truth']['disaster_type']}")
            print(f"  Distress: {sample['ground_truth']['distress']}")
            print(f"  Location: {sample['ground_truth']['location']}")
            print(f"  Action: {sample['ground_truth']['action']}")
            
            print("\nPrediction:")
            print(f"  Disaster Type: {sample['prediction']['disaster_type']}")
            print(f"  Distress: {sample['prediction']['distress']}")
            print(f"  Location: {sample['prediction']['location']}")
            print(f"  Action: {sample['prediction']['action']}")
            
            print("\nRetrieved Documents:")
            if 'retrieved_docs' in sample['full_prediction']:
                for j, doc in enumerate(sample['full_prediction']['retrieved_docs'][:3], 1):
                    print(f"  Doc {j}: {doc[:100]}...")
            else:
                print("  No documents retrieved")
                
            if 'extracted_locations' in sample:
                print(f"\nExtracted Locations: {sample['extracted_locations']}")
                print(f"Location in Tweet: {sample['location_in_tweet']}")
            
            print("\nMetrics:")
            print(f"  Disaster Type - BLEU: {sample['metrics']['disaster_type']['bleu']:.4f}, Exact Match: {sample['metrics']['disaster_type']['exact_match']}")
            print(f"  Distress - BLEU: {sample['metrics']['distress']['bleu']:.4f}, Exact Match: {sample['metrics']['distress']['exact_match']}")
            print(f"  Location - BLEU: {sample['metrics']['location']['bleu']:.4f}, Exact Match: {sample['metrics']['location']['exact_match']}")
            print(f"  Action - BLEU: {sample['metrics']['action']['bleu']:.4f}, Exact Match: {sample['metrics']['action']['exact_match']}")
            
            print("\n" + "=" * 80 + "\n")
    
    def calculate_metrics_so_far(self):
        """Calculate metrics on current results without saving"""
        if not self.results:
            return {}
            
        results_df = pd.DataFrame(self.results)
        
        # Exact match accuracy
        disaster_type_accuracy = results_df['disaster_type_exact'].mean()
        distress_accuracy = results_df['distress_exact'].mean()
        
        metrics = {
            'disaster_type_accuracy': disaster_type_accuracy,
            'distress_accuracy': distress_accuracy
        }
            
        return metrics
        
    def calculate_aggregate_metrics(self):
        """Calculate aggregate metrics across all samples"""
        if not self.results:
            print("No results to aggregate")
            return {}
            
        results_df = pd.DataFrame(self.results)
        
        # Exact match accuracy
        disaster_type_accuracy = results_df['disaster_type_exact'].mean()
        distress_accuracy = results_df['distress_exact'].mean()
        location_accuracy = results_df['location_exact'].mean()
        action_accuracy = results_df['action_exact'].mean()
        
        # Average BLEU scores
        avg_disaster_type_bleu = results_df['disaster_type_bleu'].mean()
        avg_distress_bleu = results_df['distress_bleu'].mean()
        avg_location_bleu = results_df['location_bleu'].mean()
        avg_action_bleu = results_df['action_bleu'].mean()
        
        # Average ROUGE scores
        avg_disaster_type_rouge_1 = results_df['disaster_type_rouge_1'].mean()
        avg_disaster_type_rouge_l = results_df['disaster_type_rouge_l'].mean()
        
        avg_distress_rouge_1 = results_df['distress_rouge_1'].mean()
        avg_distress_rouge_l = results_df['distress_rouge_l'].mean()
        
        avg_location_rouge_1 = results_df['location_rouge_1'].mean()
        avg_location_rouge_l = results_df['location_rouge_l'].mean()
        
        avg_action_rouge_1 = results_df['action_rouge_1'].mean()
        avg_action_rouge_l = results_df['action_rouge_l'].mean()
        
        # Average processing time
        avg_processing_time = results_df['processing_time'].mean()
        
        # F1 scores
        f1_disaster_type = f1_score(
            results_df['gt_disaster_type'], 
            results_df['pred_disaster_type'], 
            average='weighted',
            labels=np.unique(results_df[['gt_disaster_type', 'pred_disaster_type']].values)
        )
        
        f1_distress = f1_score(
            results_df['gt_distress'], 
            results_df['pred_distress'], 
            average='weighted',
            labels=np.unique(results_df[['gt_distress', 'pred_distress']].values)
        )
        
        # Print aggregate metrics
        print("\n===== Aggregate Metrics =====")
        print(f"Disaster Type Accuracy: {disaster_type_accuracy:.4f}")
        print(f"Distress Detection Accuracy: {distress_accuracy:.4f}")
        print(f"Location Extraction Accuracy: {location_accuracy:.4f}")
        print(f"Action Recommendation Accuracy: {action_accuracy:.4f}")
        
        print(f"\nF1 Score - Disaster Type: {f1_disaster_type:.4f}")
        print(f"F1 Score - Distress Detection: {f1_distress:.4f}")
        
        print(f"\n===== BLEU Scores =====")
        print(f"Disaster Type: {avg_disaster_type_bleu:.4f}")
        print(f"Distress: {avg_distress_bleu:.4f}")
        print(f"Location: {avg_location_bleu:.4f}")
        print(f"Action: {avg_action_bleu:.4f}")
        
        print(f"\n===== ROUGE-1 Scores =====")
        print(f"Disaster Type: {avg_disaster_type_rouge_1:.4f}")
        print(f"Distress: {avg_distress_rouge_1:.4f}")
        print(f"Location: {avg_location_rouge_1:.4f}")
        print(f"Action: {avg_action_rouge_1:.4f}")
        
        print(f"\n===== ROUGE-L Scores =====")
        print(f"Disaster Type: {avg_disaster_type_rouge_l:.4f}")
        print(f"Distress: {avg_distress_rouge_l:.4f}")
        print(f"Location: {avg_location_rouge_l:.4f}")
        print(f"Action: {avg_action_rouge_l:.4f}")
        
        print(f"\nAverage Processing Time: {avg_processing_time:.2f} seconds")
        
        # Save aggregate metrics
        self.aggregate_metrics = {
            'disaster_type_accuracy': disaster_type_accuracy,
            'distress_accuracy': distress_accuracy,
            'location_accuracy': location_accuracy,
            'action_accuracy': action_accuracy,
            
            'f1_disaster_type': f1_disaster_type,
            'f1_distress': f1_distress,
            
            'avg_disaster_type_bleu': avg_disaster_type_bleu,
            'avg_distress_bleu': avg_distress_bleu,
            'avg_location_bleu': avg_location_bleu,
            'avg_action_bleu': avg_action_bleu,
            
            'avg_disaster_type_rouge_1': avg_disaster_type_rouge_1,
            'avg_distress_rouge_1': avg_distress_rouge_1,
            'avg_location_rouge_1': avg_location_rouge_1,
            'avg_action_rouge_1': avg_action_rouge_1,
            
            'avg_disaster_type_rouge_l': avg_disaster_type_rouge_l,
            'avg_distress_rouge_l': avg_distress_rouge_l,
            'avg_location_rouge_l': avg_location_rouge_l,
            'avg_action_rouge_l': avg_action_rouge_l,
            
            'avg_processing_time': avg_processing_time
        }
        
        return self.aggregate_metrics
        
    def save_results(self):
        """Save evaluation results to CSV and JSON"""
        if not self.results:
            print("No results to save")
            return
            
        # Save detailed results
        results_df = pd.DataFrame(self.results)
        results_df.to_csv(f"{self.output_path}.csv", index=False)
        print(f"Saved detailed results to {self.output_path}.csv")
        
        # Save sample results
        with open(f"{self.output_path}_sample_results.json", 'w') as f:
            # We need to convert some non-serializable objects
            cleaned_samples = []
            for sample in self.sample_results:
                # Make a copy to avoid modifying the original
                cleaned_sample = sample.copy()
                # Remove full prediction which might contain non-serializable objects
                if 'full_prediction' in cleaned_sample:
                    # Keep only the parts we need
                    cleaned_sample['full_prediction'] = {
                        'analysis': cleaned_sample['full_prediction']['analysis'],
                        'retrieved_docs': cleaned_sample['full_prediction'].get('retrieved_docs', [])[:3]
                    }
                cleaned_samples.append(cleaned_sample)
            json.dump(cleaned_samples, f, indent=4)
        print(f"Saved sample results to {self.output_path}_sample_results.json")
        
        # Save aggregate metrics
        if hasattr(self, 'aggregate_metrics'):
            metrics_df = pd.DataFrame([self.aggregate_metrics])
            metrics_path = f"{self.output_path}_metrics.csv"
            metrics_df.to_csv(metrics_path, index=False)
            print(f"Saved aggregate metrics to {metrics_path}")
            
            # Also save as JSON for easier reading
            metrics_json_path = f"{self.output_path}_metrics.json"
            with open(metrics_json_path, 'w') as f:
                json.dump(self.aggregate_metrics, f, indent=4)
            print(f"Saved aggregate metrics to {metrics_json_path}")
            
    def generate_detailed_classification_reports(self, results_df):
        """Generate detailed classification reports for all tasks"""
        # Dictionary to store all reports
        all_reports = {}
        
        # Disaster Type Classification Report
        print("\n===== Disaster Type Classification Report =====")
        disaster_type_report = classification_report(
            results_df['gt_disaster_type'],
            results_df['pred_disaster_type'],
            output_dict=True
        )
        disaster_type_df = pd.DataFrame(disaster_type_report).transpose()
        print(disaster_type_df.round(3))
        all_reports['disaster_type'] = disaster_type_report
        
        # Distress Detection Classification Report
        print("\n===== Distress Detection Classification Report =====")
        distress_report = classification_report(
            results_df['gt_distress'],
            results_df['pred_distress'],
            output_dict=True
        )
        distress_df = pd.DataFrame(distress_report).transpose()
        print(distress_df.round(3))
        all_reports['distress'] = distress_report
        
        # Location Extraction Classification Report
        # For location, we need to handle multiple formats - simplify to check if it matched or not
        print("\n===== Location Extraction Classification Report =====")
        # Create simplified categories for location
        results_df['gt_location_simple'] = results_df['gt_location'].apply(
            lambda x: "specified" if x != "unknown" else "unknown"
        )
        results_df['pred_location_simple'] = results_df['pred_location'].apply(
            lambda x: "specified" if x != "unknown" else "unknown"
        )
        
        location_report = classification_report(
            results_df['gt_location_simple'],
            results_df['pred_location_simple'],
            output_dict=True
        )
        location_df = pd.DataFrame(location_report).transpose()
        print(location_df.round(3))
        all_reports['location'] = location_report
        
        # Action Recommendation Classification Report
        print("\n===== Action Recommendation Classification Report =====")
        action_report = classification_report(
            results_df['gt_action'],
            results_df['pred_action'],
            output_dict=True
        )
        action_df = pd.DataFrame(action_report).transpose()
        print(action_df.round(3))
        all_reports['action'] = action_report
        
        # Save classification reports
        disaster_type_df.to_csv(f"{self.output_path}_disaster_type_report.csv")
        distress_df.to_csv(f"{self.output_path}_distress_report.csv")
        location_df.to_csv(f"{self.output_path}_location_report.csv")
        action_df.to_csv(f"{self.output_path}_action_report.csv")
        
        return all_reports
    
    def generate_metrics_table(self, results_df, metrics):
        """Generate a comprehensive metrics table"""
        # Create a table with all metrics
        metrics_table = pd.DataFrame({
            'Task': ['Disaster Type', 'Distress', 'Location', 'Action'],
            'Accuracy': [
                metrics['disaster_type_accuracy'], 
                metrics['distress_accuracy'],
                metrics['location_accuracy'],
                metrics['action_accuracy']
            ],
            'BLEU Score': [
                metrics['avg_disaster_type_bleu'],
                metrics['avg_distress_bleu'],
                metrics['avg_location_bleu'],
                metrics['avg_action_bleu']
            ],
            'ROUGE-1 Score': [
                metrics['avg_disaster_type_rouge_1'],
                metrics['avg_distress_rouge_1'],
                metrics['avg_location_rouge_1'],
                metrics['avg_action_rouge_1']
            ],
            'ROUGE-L Score': [
                metrics['avg_disaster_type_rouge_l'],
                metrics['avg_distress_rouge_l'],
                metrics['avg_location_rouge_l'],
                metrics['avg_action_rouge_l']
            ]
        })
        
        # Add F1 scores where applicable
        if 'f1_disaster_type' in metrics and 'f1_distress' in metrics:
            metrics_table['F1 Score'] = [
                metrics['f1_disaster_type'],
                metrics['f1_distress'],
                np.nan,  # No overall F1 for location
                np.nan   # No overall F1 for action
            ]
        
        # Display and save the table
        print("\n===== Comprehensive Metrics Table =====")
        print(metrics_table.round(4))
        
        # Save to CSV
        metrics_table.to_csv(f"{self.output_path}_metrics_table.csv", index=False)
        
        # Create a styled HTML version for better visualization
        styled_table = metrics_table.style.format("{:.4f}", subset=metrics_table.columns[1:])
        styled_table.to_html(f"{self.output_path}_metrics_table.html")

    def visualize_results(self, results_df, metrics):
        """Generate comprehensive visualizations for evaluation results"""
        # Set the style for plots
        sns.set(style="whitegrid")
        plt.rcParams['figure.figsize'] = (12, 8)
        
        # 1. Accuracy Metrics
        plt.figure(figsize=(10, 6))
        accuracy_metrics = [
            ('disaster_type_accuracy', 'Disaster Type'),
            ('distress_accuracy', 'Distress'),
            ('location_accuracy', 'Location'),
            ('action_accuracy', 'Action')
        ]
        
        values = [metrics[key] for key, _ in accuracy_metrics]
        labels = [label for _, label in accuracy_metrics]
        
        bars = plt.bar(labels, values, color=['#3498db', '#2ecc71', '#e74c3c', '#f39c12'])
        plt.ylim(0, 1.0)
        plt.title('Classification Accuracy by Task', fontsize=16)
        plt.ylabel('Accuracy')
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        
        # Add value labels on bars
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{height:.3f}', ha='center', fontsize=11)
        
        plt.tight_layout()
        plt.savefig(f"{self.output_path}_accuracy_metrics.png", dpi=300, bbox_inches='tight')
        plt.show()
        
        # 2. BLEU Scores by Field - Radar Chart
        plt.figure(figsize=(8, 8))
        
        # Prepare data for radar chart
        bleu_categories = ['Disaster Type', 'Distress', 'Location', 'Action']
        bleu_values = [
            metrics['avg_disaster_type_bleu'],
            metrics['avg_distress_bleu'],
            metrics['avg_location_bleu'],
            metrics['avg_action_bleu']
        ]
        
        # Create radar chart for BLEU scores
        ax = plt.subplot(111, polar=True)
        angles = np.linspace(0, 2*np.pi, len(bleu_categories), endpoint=False).tolist()
        angles += angles[:1]  # close the loop
        
        bleu_values += bleu_values[:1]  # close the loop
        
        ax.plot(angles, bleu_values, 'o-', linewidth=2, color='#3498db')
        ax.fill(angles, bleu_values, alpha=0.25, color='#3498db')
        
        ax.set_thetagrids(np.degrees(angles[:-1]), bleu_categories)
        ax.set_ylim(0, max(bleu_values) * 1.2)
        ax.set_title('BLEU Scores by Task', fontsize=16, y=1.1)
        
        # Add value labels
        for angle, value in zip(angles[:-1], bleu_values[:-1]):
            ax.text(angle, value + 0.02, f'{value:.3f}', 
                    horizontalalignment='center', verticalalignment='bottom')
        
        plt.tight_layout()
        plt.savefig(f"{self.output_path}_bleu_radar.png", dpi=300, bbox_inches='tight')
        plt.show()
        
        # 3. ROUGE Scores by Field - Radar Chart
        plt.figure(figsize=(8, 8))
        
        # Prepare data for radar chart
        rouge_categories = ['Disaster Type', 'Distress', 'Location', 'Action']
        rouge1_values = [
            metrics['avg_disaster_type_rouge_1'],
            metrics['avg_distress_rouge_1'],
            metrics['avg_location_rouge_1'],
            metrics['avg_action_rouge_1']
        ]
        
        rougeL_values = [
            metrics['avg_disaster_type_rouge_l'],
            metrics['avg_distress_rouge_l'],
            metrics['avg_location_rouge_l'],
            metrics['avg_action_rouge_l']
        ]
        
        # Create radar chart for ROUGE scores
        ax = plt.subplot(111, polar=True)
        angles = np.linspace(0, 2*np.pi, len(rouge_categories), endpoint=False).tolist()
        angles += angles[:1]  # close the loop
        
        rouge1_values += rouge1_values[:1]  # close the loop
        rougeL_values += rougeL_values[:1]  # close the loop
        
        ax.plot(angles, rouge1_values, 'o-', linewidth=2, color='#e74c3c', label='ROUGE-1')
        ax.fill(angles, rouge1_values, alpha=0.1, color='#e74c3c')
        
        ax.plot(angles, rougeL_values, 'o-', linewidth=2, color='#9b59b6', label='ROUGE-L')
        ax.fill(angles, rougeL_values, alpha=0.1, color='#9b59b6')
        
        ax.set_thetagrids(np.degrees(angles[:-1]), rouge_categories)
        ax.set_ylim(0, max(max(rouge1_values), max(rougeL_values)) * 1.2)
        ax.set_title('ROUGE Scores by Task', fontsize=16, y=1.1)
        plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
        
        # Add value labels for ROUGE-1
        for angle, value in zip(angles[:-1], rouge1_values[:-1]):
            ax.text(angle, value + 0.02, f'{value:.3f}', 
                    horizontalalignment='center', verticalalignment='bottom',
                    color='#e74c3c')
        
        # Add value labels for ROUGE-L
        for angle, value in zip(angles[:-1], rougeL_values[:-1]):
            ax.text(angle, value + 0.02, f'{value:.3f}', 
                    horizontalalignment='center', verticalalignment='top',
                    color='#9b59b6')
        
        plt.tight_layout()
        plt.savefig(f"{self.output_path}_rouge_radar.png", dpi=300, bbox_inches='tight')
        plt.show()
        
        # 4. Confusion Matrices for all tasks
        
        # Disaster Type Confusion Matrix
        plt.figure(figsize=(10, 8))
        unique_disaster_types = sorted(list(set(
            results_df['gt_disaster_type'].dropna().unique().tolist() + 
            results_df['pred_disaster_type'].dropna().unique().tolist()
        )))
        
        disaster_cm = confusion_matrix(
            results_df['gt_disaster_type'], 
            results_df['pred_disaster_type'],
            labels=unique_disaster_types
        )
        
        sns.heatmap(disaster_cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=unique_disaster_types, yticklabels=unique_disaster_types)
        plt.title('Disaster Type Confusion Matrix', fontsize=16)
        plt.xlabel('Predicted')
        plt.ylabel('Ground Truth')
        plt.tight_layout()
        plt.savefig(f"{self.output_path}_disaster_type_cm.png", dpi=300, bbox_inches='tight')
        plt.show()
        
        # Distress Confusion Matrix
        plt.figure(figsize=(8, 6))
        unique_distress = sorted(list(set(
            results_df['gt_distress'].dropna().unique().tolist() + 
            results_df['pred_distress'].dropna().unique().tolist()
        )))
        
        distress_cm = confusion_matrix(
            results_df['gt_distress'], 
            results_df['pred_distress'],
            labels=unique_distress
        )
        
        sns.heatmap(distress_cm, annot=True, fmt='d', cmap='Greens',
                    xticklabels=unique_distress, yticklabels=unique_distress)
        plt.title('Distress Detection Confusion Matrix', fontsize=16)
        plt.xlabel('Predicted')
        plt.ylabel('Ground Truth')
        plt.tight_layout()
        plt.savefig(f"{self.output_path}_distress_cm.png", dpi=300, bbox_inches='tight')
        plt.show()
        
        # Location Confusion Matrix (simplified)
        plt.figure(figsize=(8, 6))
        unique_locations = ['unknown', 'specified']
        
        location_cm = confusion_matrix(
            results_df['gt_location_simple'],
            results_df['pred_location_simple'],
            labels=unique_locations
        )
        
        sns.heatmap(location_cm, annot=True, fmt='d', cmap='Oranges',
                    xticklabels=unique_locations, yticklabels=unique_locations)
        plt.title('Location Extraction Confusion Matrix', fontsize=16)
        plt.xlabel('Predicted')
        plt.ylabel('Ground Truth')
        plt.tight_layout()
        plt.savefig(f"{self.output_path}_location_cm.png", dpi=300, bbox_inches='tight')
        plt.show()
        
        # Action Confusion Matrix
        plt.figure(figsize=(12, 10))
        unique_actions = sorted(list(set(
            results_df['gt_action'].dropna().unique().tolist() + 
            results_df['pred_action'].dropna().unique().tolist()
        )))
        
        # If there are too many unique actions, limit to the most common ones
        if len(unique_actions) > 10:
            action_counts = results_df['gt_action'].value_counts()
            top_actions = action_counts.nlargest(10).index.tolist()
            unique_actions = sorted(top_actions)
        
        action_cm = confusion_matrix(
            results_df['gt_action'], 
            results_df['pred_action'],
            labels=unique_actions
        )
        
        plt.figure(figsize=(12, 10))
        sns.heatmap(action_cm, annot=True, fmt='d', cmap='Purples',
                    xticklabels=unique_actions, yticklabels=unique_actions)
        plt.title('Action Recommendation Confusion Matrix', fontsize=16)
        plt.xlabel('Predicted')
        plt.ylabel('Ground Truth')
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.savefig(f"{self.output_path}_action_cm.png", dpi=300, bbox_inches='tight')
        plt.show()
        
        # 5. Processing Time Distribution
        plt.figure(figsize=(10, 6))
        sns.histplot(results_df['processing_time'], bins=20, kde=True, color='#3498db')
        plt.axvline(metrics['avg_processing_time'], color='red', linestyle='dashed', linewidth=2)
        plt.text(metrics['avg_processing_time'] * 1.05, plt.ylim()[1] * 0.9, 
                f'Mean: {metrics["avg_processing_time"]:.2f}s', color='red', fontsize=12)
        plt.title('Processing Time Distribution', fontsize=16)
        plt.xlabel('Processing Time (seconds)')
        plt.ylabel('Frequency')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig(f"{self.output_path}_processing_time.png", dpi=300, bbox_inches='tight')
        plt.show()


def run_comprehensive_evaluation(rag_system, output_prefix="llama_evaluation", max_samples=None, display_samples=5):
    """Run a comprehensive evaluation of the RAG system"""
    print("Starting comprehensive evaluation of RAG system...")
    
    # Create evaluator
    evaluator = RAGSystemEvaluator(
        rag_system=rag_system,
        sample_dataset_path="sample_dataset_balanced.csv",
        output_path=output_prefix
    )
    
    # Run evaluation
    results_df, metrics = evaluator.evaluate_all_samples(max_samples)
    
    # Display sample results
    evaluator.display_sample_results(display_samples)
    
    # Generate detailed classification reports
    all_reports = evaluator.generate_detailed_classification_reports(results_df)
    
    # Generate metrics table
    metrics_table = evaluator.generate_metrics_table(results_df, metrics)
    
    # Visualize results
    evaluator.visualize_results(results_df, metrics)
    
    print("\nComprehensive evaluation complete!")
    return results_df, metrics, all_reports, metrics_table

# Import NLTK data for BLEU score calculation
import nltk
nltk.download('punkt')

# Now run the evaluation with your existing RAG system
# Use max_samples parameter to limit evaluation to fewer examples during testing
results_df, metrics, all_reports, metrics_table = run_comprehensive_evaluation(
    rag_system=rag_system,  # This is your existing RAG system that should be already defined
    output_prefix="llama_evaluation",
    max_samples=None,        # Set to None to evaluate all samples or a number for quick testing
    display_samples=5        # Number of sample results to display
)
```
